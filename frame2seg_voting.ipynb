```python
import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.stats import itemfreq
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

### Prepare training data

```python
fold1 = np.load('./cv/fold1_unshape.npy')
fold2 = np.load('./cv/fold2_unshape.npy')
fold3 = np.load('./cv/fold3_unshape.npy')
fold4 = np.load('./cv/fold4_unshape.npy')
```

```python
train = fold1
train = np.append(train, fold2, axis=0)
train = np.append(train, fold3, axis=0)
#train = np.append(train, fold4, axis=0)
```

```python
pd.DataFrame(train)
```

```python
fold1_class = np.load('./cv/fold1_class_unshape.npy')
fold2_class = np.load('./cv/fold2_class_unshape.npy')
fold3_class = np.load('./cv/fold3_class_unshape.npy')
fold4_class = np.load('./cv/fold4_class_unshape.npy')
```

```python
train_class = fold1_class
train_class = np.append(train_class, fold2_class, axis=0)
train_class = np.append(train_class, fold3_class, axis=0)
#train_class = np.append(train_class, fold4_class, axis=0)
```

```python
pd.DataFrame(train_class)
```

### Build model

#### Here, I didn't use cv to evalute the model. I use the first three folds as training data, and the other fold(4th fold) as testing data.

```python
## bulid model
clf_nb = GaussianNB(priors=None)
clf_nb.fit(train, train_class)
```

### Test on fold 4

```python
with open('./TUT-acoustic-scenes-2017-development/evaluation_setup/fold4_evaluate_revise.txt') as f:
    lines = f.readlines()
```

```python
class2num = {"class":     {"bus": 0, "cafe/restaurant": 1, "car": 2, "city_center": 3, "forest_path": 4, 
                              "grocery_store": 5, "home": 6, "beach": 7, "library": 8, "metro_station": 9, 
                              "office": 10, "residential_area": 11, "train": 12, "tram": 13, "park": 14}}
```

```python
y_true = np.empty(shape=[0,1])
y_pred = np.empty(shape=[0,1])
```

```python
for line in lines:
    description = line.split('\t')
    
    ## decoding class to string
    description[1] = description[1].replace('\n', '')
    
    ## convert string to numerical value
    description[1] = class2num["class"][description[1]]
    
    ## append true value
    y_true = np.append(y_true, description[1])
    
    ## import data
    try:
        data = np.load('./DCASE2017_development_set/mfcc_numpy/' + description[0])
    except:
        print("ERROR")
        
    ## convert to data frame
    df = pd.DataFrame(data)
    df = df.T
    
    ## prediction
    predicts = clf_nb.predict(df.values)
    
    ## voting
    pred_cnt = itemfreq(predicts)
    voting_result = pred_cnt[pred_cnt[:,1].argsort()][-1,0]
    
    ## append voting result
    y_pred = np.append(y_pred, voting_result)
```

```python
accuracy_score(y_true, y_pred)
```

### Final model

#### In the final model, I append the 4-th fold into training data. Then predict the new data

```python
## add fold 4 in to train data
train_final = np.append(train, fold4, axis=0)

## add classes into train label
train_class_final = np.append(train_class, fold4_class, axis=0)
```

```python
## bulid model
clf_nb_final = GaussianNB(priors=None)
clf_nb_final.fit(train_final, train_class_final)
```
