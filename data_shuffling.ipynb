# Data Pre-processing

```python
# import dependencies
import os
import numpy as np
import pandas as pd

from scipy.io.arff import loadarff
from os import listdir
```

```python
FOLDER = 'mfcc_weka/'
weka_files = os.listdir(FOLDER)

data_df = []
for file in weka_files:
    print('Loading data from file ', file)
    data = loadarff(open(FOLDER + file, 'r'))[0]
    try:
        data_df = np.concatenate([data_df, data], axis=0)
    except TypeError:
        print('Error')
        data_df = data
data_df
```

```python
data = pd.DataFrame(data_df)
data
```

```python
np.random.shuffle(data_df)
data_shuf = pd.DataFrame(data_df)
data_shuf
```

```python
os.mkdir('data_processed')
for i in range(24):
    np.save(
        'data_processed/' + 'chunk' + str(i+1) + '.npy', 
        data_df[i * 100000:(i + 1) * 100000]
    )
```
