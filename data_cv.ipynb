```python
# import dependencies
import os
import numpy as np
import pandas as pd

from scipy.io.arff import loadarff
from os import listdir
```

### type fold name

```python
selected_fold = 'fold4'
```

```python
with open('./cv/'+selected_fold+'_file_names.txt') as f:
    lines = f.readlines()
```

```python
class2num = {"class":     {"bus": 0, "cafe/restaurant": 1, "car": 2, "city_center": 3, "forest_path": 4, 
                              "grocery_store": 5, "home": 6, "beach": 7, "library": 8, "metro_station": 9, 
                              "office": 10, "residential_area": 11, "train": 12, "tram": 13, "park": 14}}
```

```python
fold1 = np.empty(shape=[0,501*60])
fold1_class = np.empty(shape=[0,1])

cnt=0
for line in lines:
    line = line.replace("\n","")
    file_name = './DCASE2017_development_set/mfcc_weka/' + line + '.arff'
    print("import file:\t" + line)
    
    # load weka files
    data_arff = loadarff(open(file_name, 'r'))
    data_arff_np = np.array(data_arff[0])
    
    df = pd.DataFrame(data_arff_np)
    
    ## decoding class to string
    df['class'] = df['class'].str.decode("utf-8")
    
    ## convert string to numerical value
    df.replace(class2num, inplace=True)
    
    ## split attributes and classes
    df_attribute = df.drop('class', axis=1).values
    df_class = df['class'].values[0]
    
    ## reshape the array
    df_attribute = df_attribute.reshape(1,501*60)
    
    fold1 = np.append(fold1, df_attribute, axis=0)
    fold1_class = np.append(fold1_class, df_class)
    cnt+=1
    
print("In total\t" + str(cnt) + "\tfiles")
```
