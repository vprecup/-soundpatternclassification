```python
# import dependencies
import os
import re
import numpy as np

import converter as conv

from os import listdir
```

# Iterate over the fold config files
And collect the data

```python
def collect_data(filelist):
    data = np.empty([filelist.shape[0], 501, 60])
    for index, file in enumerate(filelist):
        correct_filename = 'mfcc_numpy/{}.npy'.format(re.match(r'audio/(.+).wav', file).group(1))
        print('Loading data from... {}'.format(correct_filename))
        data[index] = np.load(correct_filename).T
    return data
```

```python
if not os.path.exists('fold_arrays'):
    folds_dir = os.mkdir('fold_arrays')

for i in range(1, 5, 1):
    config = np.genfromtxt('mfcc_numpy/fold{}_evaluate.txt'.format(i), dtype=np.str, delimiter='\t')
    data = collect_data(config[:,0])
    np.save('fold_arrays/fold{}_x.npy'.format(i), data)
    labels = conv.labels_to_numeric(config[:,1], decode=False)
    np.save('fold_arrays/fold{}_y.npy'.format(i), labels)
```
